```python
from collections import Counter
import nltk
import pandas as pd
import re as regex
import numpy as np
import plotly
from plotly import graph_objs
from sklearn.metrics import f1_score, precision_score, recall_score, accuracy_score
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV, RandomizedSearchCV
from time import time
import plotly.graph_objs as go
# plotly configuration
plotly.offline.init_notebook_mode()
```

```python
class Load_Tweets():
    data = []
    processed_data = []
    wordlist = []

    data_model = None
    data_labels = None
    is_testing = False

    def load(self, csv_file, is_testing_set=False, from_cached=None):
        if from_cached is not None:
            self.data_model = pd.read_csv(from_cached)
            return

        self.is_testing = is_testing_set

        if not is_testing_set:
            self.data = pd.read_csv(csv_file, header=0, names=["id", "text", "category"],
                                    dtype={"id": "str", "text": "str", "category": "str"},encoding='ANSI')
            self.data = self.data[self.data["category"].isin(["positive", "negative", "neutral"])]
        else:
            #self.data = pd.read_csv(csv_file, header=0, names=["id", "text"], dtype={"id": "str", "text": "str"},
             #                       nrows=4000)
            self.data = pd.read_csv(csv_file, header=0, names=["id", "text"], dtype={"id": "str", "text": "str"},encoding='ANSI')                                    
            not_null_text = 1 ^ pd.isnull(self.data["text"])
            not_null_id = 1 ^ pd.isnull(self.data["id"])
            self.data = self.data.loc[not_null_id & not_null_text, :]

        self.processed_data = self.data
        self.wordlist = []
        self.data_model = None
        self.data_labels = None
```

```python
data = Load_Tweets()
data.load("C:/Users/stark/Downloads/tweetScripts/08052018/train.csv")
data.processed_data.head(5)
```

```python
class TokenizeStem_Tweets(Load_Tweets):
    def __init__(self, previous):
        self.processed_data = previous.processed_data

    def tokenize(self, tokenizer=nltk.word_tokenize):
        def tokenize_row(row):
            row["text"] = tokenizer(row["text"])
            row["text"] = [] + row["text"]
            return row

        self.processed_data = self.processed_data.apply(tokenize_row, axis=1)
```

```python
class WordsList_Tweets(TokenizeStem_Tweets):
    def __init__(self, previous):
        self.processed_data = previous.processed_data

    whitelist = ["n't", "not"]
    wordlist = []

    def generate_wordlist(self, is_testing, min_occurrences=3, max_occurences=500, stopwords=nltk.corpus.stopwords.words("english"),
                       whitelist=None):
        self.wordlist = []
        whitelist = self.whitelist if whitelist is None else whitelist
        import os
        if not is_testing:
            if os.path.isfile("C:/Users/stark/Downloads/tweetScripts/08052018/wordlist.csv"):
                word_df = pd.read_csv("C:/Users/stark/Downloads/tweetScripts/08052018/wordlist.csv",encoding='ANSI')
                word_df = word_df[word_df["occurrences"] > min_occurrences]
                self.wordlist = list(word_df.loc[:, "word"])
                return

            words = Counter()
            for ids in self.processed_data.index:
                words.update(self.processed_data.loc[ids, "text"])

            for ids, stop_word in enumerate(stopwords):
                if stop_word not in whitelist:
                    del words[stop_word]

            word_df = pd.DataFrame(
                data={"word": [k for k, v in words.most_common() if min_occurrences < v < max_occurences],
                  "occurrences": [v for k, v in words.most_common() if min_occurrences < v < max_occurences]},
                columns=["word", "occurrences"])

            word_df.to_csv("C:/Users/stark/Downloads/tweetScripts/08052018/wordlist.csv", index_label="ids")
            self.wordlist = [k for k, v in words.most_common() if min_occurrences < v < max_occurences]
            
        else :
            if os.path.isfile("C:/Users/stark/Downloads/tweetScripts/08052018/wordlist_test.csv"):
                word_df = pd.read_csv("C:/Users/stark/Downloads/tweetScripts/08052018/wordlist_test.csv",encoding='ANSI')
                word_df = word_df[word_df["occurrences"] > min_occurrences]
                self.wordlist = list(word_df.loc[:, "word"])
                return

            words = Counter()
            for ids in self.processed_data.index:
                words.update(self.processed_data.loc[ids, "text"])

            for ids, stop_word in enumerate(stopwords):
                if stop_word not in whitelist:
                    del words[stop_word]

            word_df = pd.DataFrame(
                data={"word": [k for k, v in words.most_common() if min_occurrences < v < max_occurences],
                  "occurrences": [v for k, v in words.most_common() if min_occurrences < v < max_occurences]},
                columns=["word", "occurrences"])

            word_df.to_csv("C:/Users/stark/Downloads/tweetScripts/08052018/wordlist_test.csv", index_label="ids")
            self.wordlist = [k for k, v in words.most_common() if min_occurrences < v < max_occurences]
```

```python
class Bagofwords_Tweets(WordsList_Tweets):
    def __init__(self, previous):
        self.processed_data = previous.processed_data
        self.wordlist = previous.wordlist

    def develop_data_model(self,is_testing):
        label_column = []
        #if not self.is_testing:
        if not is_testing:
            label_column = ["label"]

        columns = label_column + list(
            map(lambda w: w + "_bow", self.wordlist))
        labels = []
        rows = []
        for ids in self.processed_data.index:
            current_row = []

            #if not self.is_testing:
            if not is_testing:
                # add label
                current_label = self.processed_data.loc[ids, "category"]
                labels.append(current_label)
                current_row.append(current_label)

            # add bag-of-words
            tokens = set(self.processed_data.loc[ids, "text"])
            for _, word in enumerate(self.wordlist):
                current_row.append(1 if word in tokens else 0)

            rows.append(current_row)

        self.data_model = pd.DataFrame(rows, columns=columns)
        self.data_labels = pd.Series(labels)
        return self.data_model, self.data_labels
```

```python
def test_classifier(X_train, y_train, X_test, y_test, classifier,name1):
    show("")
    show("===============================================")
    classifier_name = str(type(classifier).__name__)
    show("Testing " + classifier_name)
    now = time()
    list_of_labels = sorted(list(set(y_train)))
    model = classifier.fit(X_train, y_train)
    name=name1
    if(name == "naivebayes"):        
        pkl_file = open(name + '.pkl', 'wb')
        pckl.dump(model, pkl_file)
        pkl_file.close()
    show("Learing time {0}s".format(time() - now))
    now = time()
    predictions = model.predict(X_test)
    show("Predicting time {0}s".format(time() - now))
    precision = precision_score(y_test, predictions, average=None, pos_label=None, labels=list_of_labels)
    recall = recall_score(y_test, predictions, average=None, pos_label=None, labels=list_of_labels)
    accuracy = accuracy_score(y_test, predictions)
    f1 = f1_score(y_test, predictions, average=None, pos_label=None, labels=list_of_labels)
    show("=================== Performance Evaluation ===================")
    show("            Negative     Neutral     Positive")
    show("F1       " + str(f1))
    show("Precision" + str(precision))
    show("Recall   " + str(recall))
    show("Accuracy " + str(accuracy))
    show("===============================================")

    return precision, recall, accuracy, f1

def show(x):
    print(x)
```

```python
def cv(classifier, X_train, y_train):
    show("===============================================")
    classifier_name = str(type(classifier).__name__)
    now = time()
    show("Crossvalidating " + classifier_name + "...")
    accuracy = [cross_val_score(classifier, X_train, y_train, cv=10, n_jobs=-1)]
    show("Crosvalidation completed in {0}s".format(time() - now))
    show("Accuracy: " + str(accuracy[0]))
    show("Average accuracy: " + str(np.array(accuracy[0]).mean()))
    show("===============================================")
    return accuracy
```

```python
def tailoredrandomforest_classifier(X_train, y_train, X_test, y_test, classifier,name1):
    show("")
    show("===============================================")
    classifier_name = str(type(classifier).__name__)
    show("Testing Tailored Random Forest")
    now = time()
    list_of_labels = sorted(list(set(y_train)))
    model = classifier.fit(X_train, y_train)
    
    name=name1
    #if(name == "naivebayes"):        
    pkl_file = open(name + '.pkl', 'wb')
    pckl.dump(model, pkl_file)
    pkl_file.close()
    
    show("Learing time {0}s".format(time() - now))
    now = time()
    predictions = model.predict(X_test)
    show("Predicting time {0}s".format(time() - now))

    precision = precision_score(y_test, predictions, average=None, pos_label=None, labels=list_of_labels)
    recall = recall_score(y_test, predictions, average=None, pos_label=None, labels=list_of_labels)
    accuracy = accuracy_score(y_test, predictions)
    f1 = f1_score(y_test, predictions, average=None, pos_label=None, labels=list_of_labels)
    show("=================== Performance Evaluation ===================")
    show("            Negative     Neutral     Positive")
    show("F1       " + str(f1))
    show("Precision" + str(precision))
    show("Recall   " + str(recall))
    show("Accuracy " + str(accuracy))
    show("===============================================")

    return precision, recall, accuracy, f1

def show(x):
    print(x)
```
